```python
import matplotlib.pyplot as plt
from skimage import data
from skimage.transform import swirl
import cv2

image = data.checkerboard()
swirled = swirl(image, rotation=0, strength=10, radius=60)

img2 = cv2.imread('Monkey-Puzzle-Tree.jpg')
swirled2 = swirl(img2, rotation=45, strength=50, radius=240)

fig, (ax0, ax1) = plt.subplots(nrows=1, ncols=2, figsize=(8, 3), sharex=True, sharey=True)
fig, (ax2, ax3) = plt.subplots(nrows=1, ncols=2, figsize=(8, 3), sharex=True, sharey=True)

ax0.imshow(image, cmap=plt.cm.gray)
ax0.axis('on')
ax1.imshow(swirled, cmap=plt.cm.gray)
ax1.axis('on')
ax2.imshow(img2)
ax2.axis('on')
ax3.imshow(swirled2)
ax3.axis('on')
plt.show()
```
