```python
import cv2 as cv 
import numpy as np
import matplotlib.pyplot as plt

image = cv.imread("GhostTeam_CODM.jpg")

h, w = image.shape[:2]

rotation_matrix = cv.getRotationMatrix2D((w/2,h/2), -180, 0.75)

rotated_image = cv.warpAffine(image, rotation_matrix, (w, h))

plt.subplot(121), plt.imshow(cv.cvtColor(image, cv.COLOR_BGR2RGB)), plt.title('Normal Image')
plt.subplot(122), plt.imshow(cv.cvtColor(rotated_image, cv.COLOR_BGR2RGB)) 
plt.title("Rotation") 
plt.show()
```
