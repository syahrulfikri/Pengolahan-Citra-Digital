```python
import cv2 as cv 
import numpy as np
import matplotlib.pyplot as plt

image = cv.imread("Ghost_CODM.jpg") 

h, w = image.shape[:2]
half_height, half_width = h//4, w//5 
transition_matrix = np.float32([[1, 0, half_width],[0, 1, half_height]])

img_transition = cv.warpAffine(image, transition_matrix, (w, h))

plt.subplot(121), plt.imshow(cv.cvtColor(image, cv.COLOR_BGR2RGB)), plt.title('Normal Image')
plt.subplot(122), plt.imshow(cv.cvtColor(img_transition, cv.COLOR_BGR2RGB)) 
plt.title("Translation") 
plt.show()
```
